````python
#First graph

import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import pandas as pd

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Remove rows with missing data for flipper_length_mm
penguins = penguins.dropna(subset=["flipper_length_mm"])

# Create histograms for each species
fig = px.histogram(penguins, x="flipper_length_mm", color="species", marginal="box", nbins=30, title="Flipper Length of Penguins by Species")

# Add vertical lines and rectangles for each species
for species in penguins["species"].unique():
    # Subset data for the species
    species_data = penguins[penguins["species"] == species]["flipper_length_mm"]
    
    # Calculate statistics
    mean_fl = species_data.mean()
    median_fl = species_data.median()
    min_fl = species_data.min()
    max_fl = species_data.max()
    q1_fl = species_data.quantile(0.25)
    q3_fl = species_data.quantile(0.75)
    std_fl = species_data.std()

    # Add vertical lines for mean and median
    fig.add_vline(x=mean_fl, line=dict(color="blue", width=2, dash="dash"), annotation_text=f"{species} mean", annotation_position="top left")
    fig.add_vline(x=median_fl, line=dict(color="green", width=2, dash="dot"), annotation_text=f"{species} median", annotation_position="top right")

    # Add vertical rectangles for range, IQR, and two standard deviations from the mean
    # Range
    fig.add_vrect(x0=min_fl, x1=max_fl, line_width=0, fillcolor="red", opacity=0.1, annotation_text=f"{species} Range", annotation_position="top left")
    
    # IQR
    fig.add_vrect(x0=q1_fl, x1=q3_fl, line_width=0, fillcolor="orange", opacity=0.2, annotation_text=f"{species} IQR", annotation_position="top left")
    
    # Two standard deviations from the mean
    fig.add_vrect(x0=mean_fl - 2 * std_fl, x1=mean_fl + 2 * std_fl, line_width=0, fillcolor="purple", opacity=0.1, annotation_text=f"{species} 2 Std Dev", annotation_position="top left")

# Show the plot
fig.show(renderer="png")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Remove rows with missing data for flipper_length_mm
penguins = penguins.dropna(subset=["flipper_length_mm"])

# Set up the plotting area: 1 row with 3 plots (one per species)
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Color palette for species
species_colors = {
    'Adelie': 'blue',
    'Chinstrap': 'green',
    'Gentoo': 'purple'
}

# Loop through each species to create the KDE plots and annotate statistics
for i, species in enumerate(penguins["species"].unique()):
    # Subset data for the species
    species_data = penguins[penguins["species"] == species]["flipper_length_mm"]
    
    # Calculate statistics
    mean_fl = species_data.mean()
    median_fl = species_data.median()
    min_fl = species_data.min()
    max_fl = species_data.max()
    q1_fl = species_data.quantile(0.25)
    q3_fl = species_data.quantile(0.75)
    std_fl = species_data.std()
    
    # Plot KDE for flipper_length_mm
    sns.kdeplot(species_data, ax=axes[i], color=species_colors[species], fill=True, alpha=0.5)
    
    # Add vertical lines for mean and median
    axes[i].axvline(mean_fl, color='darkblue', linestyle='--', label='Mean')
    axes[i].axvline(median_fl, color='darkgreen', linestyle=':', label='Median')
    
    # Add shaded areas for range, IQR, and 2 standard deviations
    # Range
    axes[i].axvspan(min_fl, max_fl, color='red', alpha=0.1, label='Range')
    # IQR
    axes[i].axvspan(q1_fl, q3_fl, color='orange', alpha=0.2, label='IQR')
    # Two standard deviations from the mean
    axes[i].axvspan(mean_fl - 2 * std_fl, mean_fl + 2 * std_fl, color='purple', alpha=0.1, label='2 Std Dev')
    
    # Add titles and labels
    axes[i].set_title(f"{species} Penguins", fontsize=16)
    axes[i].set_xlabel('Flipper Length (mm)', fontsize=12)
    
    # Legend handling
    if i == 0:
        axes[i].legend(loc="upper left", fontsize=10)

# Add shared y-axis label
fig.text(0.04, 0.5, 'Density', va='center', ha='center', rotation='vertical', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show(renderer="png")
````
