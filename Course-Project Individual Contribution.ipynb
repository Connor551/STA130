__Question__: 

Does having a special person to share joys and sorrows with affect how satisfied someone is with the amount of friends they have?

Data: 

I will be using the variables:
1. PSYCH_zimet_multidimensional_social_support_scale_joys - 'There is a special person with whom I can share my joys and sorrows'

__Description__: This variable indicates whether an individual has a special person with whom they can share their personal joys and sorrows. It is a measure of perceived emotional support.

__Possible answers__: 'Very Strongly Disagree', 'Strongly Disagree', 'Disagree', 'Neither Agree Nor Disagree', 'Agree', 'Strongly Agree', 'Very Strongly Agree'.

2. CONNECTION_satisfied_num_friends - 'Are you satisfied with the number of friends you have now?' (either Yes or No)

__Description__: This variable reflects whether an individual is satisfied with the current number of friends they have, with responses recorded as "Yes" or "No."

__Possible answers__: 'No, I want to have FEWER friends', 'No, I want to have MORE friends', 'Yes'.

Code below is deleting NULL values

Being able to notice that gives us an idea that the population may also have the same conclusions, but to have a better idea, we need to use bootstrapping to calculate the average answers for each category within 'PSYCH_zimet_multidimensional_social_support_scale_joys'.


__Bootstrapping__: Below I am bootstrapping the mean by setting each responce to a number to calculate a 95% interval to estimate the population mean using the given data for each category within CONNECTION_satisfied_num_friends.
(Ignore the red parts)

```python
# Import necessary libraries
import pandas as pd

# Assuming you already have a DataFrame called 'df' that contains the two columns:
# 'Special person to share my jows and sorrows' and 'Satisfied with number of friends'

# Filter the dataframe into three sections based on 'Satisfied with number of friends' column
df_fewer = df[df['Satisfied with number of friends'] == 'No, I want to have FEWER friends.']
df_more = df[df['Satisfied with number of friends'] == 'No, I want to have MORE friends.']
df_yes = df[df['Satisfied with number of friends'] == 'Yes']

# Print the value counts for the 'Special person to share my jows and sorrows' column for each category

print("Value counts for 'No, I want to have FEWER friends.':")
print(df_fewer['Special person to share my jows and sorrows'].value_counts())
print("\n")

print("Value counts for 'No, I want to have MORE friends.':")
print(df_more['Special person to share my jows and sorrows'].value_counts())
print("\n")

print("Value counts for 'Yes':")
print(df_yes['Special person to share my jows and sorrows'].value_counts())
```

```python
import numpy as np
import pandas as pd

# Create a dictionary to map responses to numeric values
response_map = {
    'Very Strongly Disagree': 1,
    'Strongly Disagree': 2,
    'Disagree': 3,
    'Neither Agree Nor Disagree': 4,
    'Agree': 5,
    'Strongly Agree': 6,
    'Very Strongly Agree': 7
}

# Function to bootstrap and calculate the average
def bootstrap_average(df, response_column, num_samples=1000):
    # Map responses to numeric values
    df['response_numeric'] = df[response_column].map(response_map)
    # List to store bootstrap sample means
    bootstrap_means = []
    
    # Bootstrap resampling
    for _ in range(num_samples):
        # Resample with replacement
        sample = df['response_numeric'].sample(frac=1, replace=True)
        # Calculate the mean of the resampled sample
        bootstrap_means.append(sample.mean())
    
    # Calculate the 95% confidence interval
    lower_bound = np.percentile(bootstrap_means, 2.5)
    upper_bound = np.percentile(bootstrap_means, 97.5)
    mean_estimate = np.mean(bootstrap_means)
    
    return mean_estimate, lower_bound, upper_bound

# Perform bootstrapping for each category
print("Bootstrapped average response for 'No, I want FEWER friends':")
mean_fewer, lower_fewer, upper_fewer = bootstrap_average(df_fewer, 'Special person to share my jows and sorrows')
print(f"Mean: {mean_fewer:.2f}, 95% CI: ({lower_fewer:.2f}, {upper_fewer:.2f})\n")

print("Bootstrapped average response for 'No, I want MORE friends':")
mean_more, lower_more, upper_more = bootstrap_average(df_more, 'Special person to share my jows and sorrows')
print(f"Mean: {mean_more:.2f}, 95% CI: ({lower_more:.2f}, {upper_more:.2f})\n")

print("Bootstrapped average response for 'Yes':")
mean_yes, lower_yes, upper_yes = bootstrap_average(df_yes, 'Special person to share my jows and sorrows')
print(f"Mean: {mean_yes:.2f}, 95% CI: ({lower_yes:.2f}, {upper_yes:.2f})\n")
```

With these numbers we can conclude:

For 'No, I want FEWER friends':
95% CI is between 'Neither Agree Nor Disagree' and 'Agree'. But the interval is slightly closer to 'Agree'

For 'No, I want MORE friends'
95% CI is 'Neither Agree Nor Disagree' and 'Agree'. But the interval is much closer to 'Agree'

For 'Yes':
95% CI is between 'Agree' and 'Strongly Agree'. But from the number we can see the interval is a bit closer to average.

From these results we can estimate for the population that people who feel stronger about having someone to talk to about their jows and sorrows will ussually feel they have the right number of friends. While people who feel like they don't have that special person to talk to about their jows and sorrows, more likely feel like they dont have the right number of friends.
